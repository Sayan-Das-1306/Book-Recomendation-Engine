```python
book_pivot=final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot.shape
```

```python
book_pivot.head()
```

```python
book_pivot=book_pivot.fillna(0)
```

```python
book_pivot.shape
```

```python
from scipy.sparse import csr_matrix
book_matrix=csr_matrix(book_pivot)
```

```python
book_matrix
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_matrix)
```

```python
book_pivot.index[237]
```

```python
book_pivot.iloc[237,:].values
```

```python
book_pivot.iloc[237,:].values.shape
```

```python
book_pivot.iloc[237,:].values.reshape(1,-1).shape
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=5)
```

```python
suggestion
```

```python
book_pivot.index[184]
```

```python
for i in suggestion[0]:
    print(book_pivot.index[i])
```

```python
np.where(book_pivot.index=='Exclusive')[0][0]
```

```python
def recommend(bookname):
    b_id=np.where(book_pivot.index==bookname)[0][0]
    distance,suggestion=model.kneighbors(book_pivot.iloc[b_id,:].values.reshape(1,-1),n_neighbors=5)
    for i in suggestion[0]:
        print(book_pivot.index[i])
```

```python
recommend('The Cradle Will Fall')
```

```python
recommend('Harry Potter and the Chamber of Secrets (Book 2)')
```
